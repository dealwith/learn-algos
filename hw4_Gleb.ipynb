–ï—Å—Ç—å n –∫–æ—Ä–æ–≤, —É –∫–∞–∂–¥–æ–π m –∫–∞—Ä—Ç –Ω–æ–º–∏–Ω–∞–ª–æ–º –æ—Ç 0 –¥–æ n * m - 1, –ø—Ä–∏ —ç—Ç–æ–º –∫–∞—Ä—Ç—ã –º–æ–≥—É—Ç –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è. –í –∫–∞–∫–æ–º –ø–æ—Ä—è–¥–∫–µ –∫–æ—Ä–æ–≤—ã –¥–æ–ª–∂–Ω—ã —Ö–æ–¥–∏—Ç—å, —á—Ç–æ–±—ã –≤—Å–µ –∏–∑ –Ω–∏—Ö –≤ –∏—Ç–æ–≥–µ –æ—Å—Ç–∞–ª–∏—Å—å –±–µ–∑ –∫–∞—Ä—Ç? –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –Ω–∞ –≤—Ö–æ–¥ –ª–∏—Å—Ç –ª–∏—Å—Ç–æ–≤ –∏ –≤—ã–¥–∞—é—â—É—é –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –ª–∏—Å—Ç - –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∫–æ—Ä–æ–≤. –ü—Ä–∏–¥—É–º–∞–π—Ç–µ —Ç–µ—Å—Ç—ã –¥–ª—è —Å–≤–æ–µ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –î–æ–∫–∞–∂–∏—Ç–µ –µ–≥–æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å. –û—Ü–µ–Ω–∏—Ç–µ –∞—Å–∏–º–ø—Ç–æ—Ç–∏–∫—É –µ–≥–æ —Ä–∞–±–æ—Ç—ã.

### –î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞

–ù–∞—à –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç –∫–∞–∂–¥—É—é –∫–∞—Ä—Ç—É –∫–∞–∂–¥–æ–≥–æ –∏–≥—Ä–æ–∫–∞, —á—Ç–æ –¥–∞–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –∫–∞–∂–¥—ã–π –∏–≥—Ä–æ–∫ –±—É–¥–µ—Ç —Ö–æ–¥–∏—Ç—å –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –∫–∞—Ä—Ç. –¢–∞–∫ –∫–∞–∫ –∫–∞—Ä—Ç—ã –º–æ–≥—É—Ç –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è, —è –ø—Ä–∏–Ω—è–ª —Ä–µ—à–µ–Ω–∏–µ –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å –ø–æ—Ä—è–¥–æ–∫ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –≤—Å–µ—Ö –∫–∞—Ä—Ç –∏–≥—Ä–æ–∫–∞. –§—É–Ω–∫—Ü–∏—è `can_complete_game` –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ –ø—Ä–æ–≤–µ—Ä–∫—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞.

### –ê—Å–∏–º–ø—Ç–æ—Ç–∏–∫–∞
1. –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∫–∞—Ä—Ç –∫–∞–∂–¥–æ–≥–æ –∏–≥—Ä–æ–∫–∞: `O(n * m * log(m))`, n - –∏–≥—Ä–æ–∫, m - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞—Ä—Ç —É –∏–≥—Ä–æ–∫–∞
2. `can_complete_game`: `O (n*m)` —Ç–∞–∫ –∫–∞–∫ 2 —Ü–∏–∫–ª–∞
3. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ: `O(n*m)`
4. `cow_avg_cards.sort()`: `nlogn`

–ò—Ç–æ–≥: `O(n * m * log(m))`, —Ç–∞–∫ –∫–∞–∫ —ç—Ç–æ –¥–æ–º–∏–Ω–∏—Ä—É—é—â–∏–π —á–ª–µ–Ω –∏ —Ä–∞—Å—Ç–µ—Ç –±—ã—Å—Ç—Ä–µ–µ –≤—Å–µ–≥–æ

````python
def find_cow_order(cards):
    players = len(cards)
    rounds = len(cards[0])

    sorted_cards = [sorted(cow_cards) for cow_cards in cards]

    def can_complete_game(perm):
        for round in range(rounds):
            top = -1

            for cow_idx in perm:
                cow_card = sorted_cards[cow_idx-1][round]

                if cow_card <= top:
                    return False

                top = cow_card

        return True

    cow_avg_cards = []

    for i, cow_cards in enumerate(sorted_cards):
        avg_card = sum(cow_cards) / len(cow_cards)
        cow_avg_cards.append((avg_card, i+1))

    cow_avg_cards.sort()
    perm = [x[1] for x in cow_avg_cards]

    if can_complete_game(perm):
        return perm
    return [-1]

def test_find_cow_order():
    print("\nTest 1: Simple case with repeating cards")
    test1 = [
        [1, 4, 4],
        [2, 3, 5],
        [0, 3, 6]
    ]
    print("Input:", test1)
    print("Output:", find_cow_order(test1))

    # Test 2
    print("\nTest 2: Impossible case")
    test2 = [
        [0, 0, 0],
        [1, 1, 1],
        [2, 2, 2]
    ]
    print("Input:", test2)
    print("Output:", find_cow_order(test2))

    # Test 3
    print("\nTest 3: Complex case with repeating cards")
    test3 = [
        [1, 5, 9],
        [2, 6, 8],
        [3, 4, 7],
        [0, 5, 10]
    ]
    print("Input:", test3)
    print("Output:", find_cow_order(test3))

    # Test 4
    print("\nTest 4: Single cow case")
    test4 = [
        [0, 1, 2]
    ]
    print("Input:", test4)
    print("Output:", find_cow_order(test4))

    # Test 5
    print("\nTest 5: Two cows with identical cards")
    test5 = [
        [1, 1, 1],
        [1, 1, 1]
    ]
    print("Input:", test5)
    print("Output:", find_cow_order(test5))

# Run tests
test_find_cow_order()
````

````python
def solve():
    t = int(input())
    results = []

    for _ in range(t):
        l, r = map(int, input().split())
        if l == 1:
            if r == 1:
                results.append("1")
            else:
                results.append(str(r - 1))
        else:
            results.append(str(r - l))

    print("\n".join(results))

solve()
# –∑–∞–¥–∞—á–∞ —Ç–∞–∫–∂–µ —Ä–µ—à–µ–Ω–∞ –Ω–∞ codeforces
````
