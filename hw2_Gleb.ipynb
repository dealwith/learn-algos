### Задача 1

Скопируйте семинарский код поиска максимума (линейный вариант) в клетку ниже. Запустите на своем компьютере со входным массивом из 100 миллионов чисел. Найдите среднее время выполнения одной операции нахождения максимума.

Попробуйте параллельно запустить на компьютере какую-нибудь требовательную к ресурсам задачу (проигрывание видео в 4к, разархивирование чего-нибудь большого). Изменилось ли время выполнения программы?

````python
import time

def find_max(inp_arr):
    m = inp_arr[0]
    for i in range(1, len(inp_arr)):
        if inp_arr[i] > m:
            m = inp_arr[i]
    return m

million_items = range(1, 100_000_001)

start_time = time.perf_counter()

maximum_number = find_max(million_items)

end_time = time.perf_counter()

print(f"maximum number: {maximum_number}")
print(f"Time to find max: {end_time - start_time:.2f} seconds")

# time to find max - 9.2 seconds
# time to find max when 4k video is on - 9.5 seconds, time to execute changed
````

В качестве входа для проверки используйте массив `[3, 4, 2, 5]`, ответ должен быть `20`.

````python
input_arr = [3, 4, 2, 5]
max_el1 = 0
max_el2 = 0
max = 0
i = 0

while (i < len(input_arr)):
	cur_el = input_arr[i]

	if (cur_el > max_el1):
		max_el2 = max_el1
		max_el1 = cur_el
		i += 1
	elif (cur_el > max_el2):
		max_el2 = cur_el
		i += 1
	else:
		i += 1

	max = max_el1 * max_el2

print("------------------------")
print("variant 2")
print(f"max multiplied: {max}")
````

## Задача 3

Вы и Некто играете в угадайку. Некто загадал число от 0 до известного числа N. Вы пытаетесь угадать это число и задаете вопросы следующего вида. Вы называете число и спрашиваете, оно больше загаданного, меньше или равно, а Некто отвечает. Вы избрали для отгадывания стратегию бинарного поиска: спрашиваете про n/2, дальше про середину подходящего интервала и так далее.

Докажите корректность такого подхода (аналогично доказательству во второй задаче), оцените его асимптотику.

Опционально напишите на Python функцию, которая будет симулировать такую игру, а точнее ваши вопросы. Например, если N = 7, а загаданное число равно 4, то последовательность вопросов будет такой:

- 1 вопрос про 3 = 7 / 2 (с округлением вниз). Ответ на него будет >
- 2 вопрос про 5, ответ <
- 3 вопрос про 4, это =

Вход функции: 7, 4
Выход функции: 3, 5, 4

````python
import math

# 22 21
def binary_search_guesses(max_num, target):
	left = 0
	right = max_num
	guesses = []

	while(True):
		mid = (left + right) // 2 # 11; 17; 20; 21;
		guesses.append(mid) # [11]; [11, 17]; [11, 17, 20]; [11, 17, 20, 21]

		if mid == target: # false; false; false; true
			break
		elif mid < target: # true; true; true
			left = mid + 1 # left == 12; left = 18; left = 21
		else:
			right = mid - 1

	return guesses

binary_search_guesses(22, 21)
````

## Доказательство корректности
- на каждой итерации мы уменьшаем диапазон поиска вдвое
- мы используем переменные `left` для хранения индекса левого диапазона, `right` для хранения правого индекса диапозона, `guesses` для хранения списка чисел, `mid` для хранения середины диапазона чисел
- инициализация переменной начинается со стартовых значений

## Асимптотическая сложность
1. первый интервал `N/2`
2. второй интервал `N/2^2`
3. третий интервал `N/2^3`
4. далее `N/2^k`

`N/2^k = 1` => `2^k = N` => `k = log2(N)` => `O(log(N))`
